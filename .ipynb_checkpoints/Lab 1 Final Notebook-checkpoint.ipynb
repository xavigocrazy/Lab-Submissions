# Introduction
Software-defined radios (SDRs) digitize radio-frequency signals by sampling an input voltage at a fixed rate and processing the resulting discrete-time sequence in software. This allows for flexible filtering and spectral analysis but it also imposes fundamental constraints set by sampling theory, most notably the Nyquist criterion and the phenomenon of aliasing.

According to the Nyquist theorem, a signal can be perfectly reconstructed only if it is sampled at a rate greater than twice its highest frequency component. Once this condition is violated, the spectral components fold into lower frequencies, this is a phenomenon known as aliasing. Typically, aliasing is treated as an undesirable effect; however, aliasing can be intentionally exploited as a tool for frequency translation noise rejection and signal isolation.  

**Working question:** How can purposeful aliasing be used to shape noise folding and preserve a clean, interpretable signal in an SDR?

To address this question, we compare time-domain sampling behavior with frequency-domain voltage and power spectra, quantify expected versus measured alias frequencies, and investigate spectral leakage and frequency resolution due to finite-time sampling and windowing. We also analyze noise statistics and demonstrate radiometer behavior through spectral averaging, linking observed stability improvements to the expected scaling.


# Methods

### Experimental Equipment
All measurements were performed using a laboratory software-defined radio (SDR) system configured for direct digitization of baseband voltage signals. The SDR includes an analog front-end, an analog-to-digital converter (ADC), and a digital signal processing chain that delivers raw sampled voltage data to the host computer. The SDR sampling rate was configurable in software and set individually for each measurement to probe behavior below, at, and above the Nyquist limit.

A laboratory signal generator was used to produce single-frequency sinusoidal waves. The signal generator output was connected directly to the SDR input using coaxial cabling. No external analog filtering was applied beyond the SDR’s inherent front-end response.

For noise measurements, a broadband noise generator was connected to the SDR input in place of the signal generator. The noise source was operated at a fixed setting labeled **level 5**, producing stationary broadband noise over the measurement bandwidth. The exact spectral shape of the noise source was not assumed a priori and was instead characterized empirically through time- and frequency-domain analysis.

During all data acquisitions, the SDR was operated using its default digital signal chain. No custom finite impulse response (FIR) filter coefficients or additional digital filtering stages were applied, allowing aliasing, noise folding, and bandpass effects to be observed directly as a function of sampling rate.


### Data Acquisition

Single-frequency sinusoidal wave were created at a fixed amplitude of **100 mVₚₚ** and injected into the SDR input. Three signal frequencies were studied: **600 kHz**, **700 kHz**, and **800 kHz**. For each signal frequency, the sampling rate of the SDR was varied to place the signal below, at, and above the Nyquist limit for that specific tone.

For the **600 kHz** signal, data were acquired at sampling rates:
- **1.0 MHz** (below Nyquist; aliased case),
- **1.2 MHz** (Nyquist-rate sampling),
- **1.4 MHz** (above Nyquist; unaliased reference).
- **3.2 MHz** (well above Nyquist).

For the **700 kHz** signal, data were acquired at:
- **1.0 MHz** (below Nyquist),
- **1.4 MHz** (Nyquist-rate sampling),
- **2.2 MHz** (above Nyquist),
- **3.2 MHz** (well above Nyquist).

For the **800 kHz** signal, data were acquired at:
- **1.4 MHz** (below Nyquist; aliased),
- **1.6 MHz** (Nyquist-rate sampling)
- **2.2 MHz** (above Nyquist),
- **3.2 MHz** (well above Nyquist).

Each acquisition produced a multi-capture dataset stored in NumPy `.npz` format, with each capture representing a single realization of the same experiment layout. Unless otherwise noted, analysis was performed on a single representative capture after verifying consistency across captures.

---

### Noise Measurements
A broadband noise generator was used as the primary noise source and operated at a fixed setting labeled **level 5**, producing approximately stationary, zero-mean noise. The noise output was applied to one input of an analog mixer. A sinusoidal reference signal from a waveform generator was applied to the second mixer input to serve as the local oscillator (LO).

The output of the analog mixer, which contains both upper and lower frequency components produced by double-sideband (DSB) mixing, was sent directly into the SDR. The SDR was set up to use quadrature (I/Q) processing and single-sideband (SSB) operation; however, the data saved for analysis consist of real-valued voltage samples rather than full complex I/Q signals. Because of this, the two frequency sidebands cannot be separated or individually identified in the recorded data. Instead, the signals analyzed in this report represent the final baseband output of the receiver configuration rather than the internal I/Q signal used by the SDR. This matters because single-sideband processing is what prevents different frequency components from overlapping in the receiver, and without directly observing the sidebands we must rely on correct receiver configuration rather than measured image suppression.


### Digital Signal Processing and Analysis Workflow

All analysis was performed in a **Jupyter notebook** using Python. Time-domain data were loaded from `.npz` files using a custom helper package (`xavis_package`) developed for this lab. For each analyzed data segment, the mean was removed to eliminate DC offsets prior to spectral analysis.

Frequency-domain representations were computed using the discrete Fourier transform (DFT). Both **voltage spectra**, proportional to |X(f)|, and **power spectra**, proportional to |X(f)|^2, were examined. Power spectra were plotted on logarithmic scales to resolve broadband noise behavior and weak spectral features. One-sided spectra were used when comparing signal and noise power within the observable bandwidth.

To reduce spectral leakage from finite-length data segments, windowing (Hann window) was applied where explicitly noted. Window functions were RMS-normalized to preserve relative power levels between windowed and unwindowed spectra. Zero-padding was used in some cases to improve visual smoothness of spectral plots but did not alter the underlying frequency resolution.

Dominant spectral peaks were identified numerically and compared against analytically predicted alias frequencies computed from the sampling rate and input tone frequency. These comparisons were used to validate deterministic aliasing behavior across Nyquist zones.

---

### Noise Statistics and Radiometer Analysis

Noise properties were examined in both the time and frequency domains. Time-domain noise distributions were compared to Gaussian models using histograms and residual analysis. In the frequency domain, the real and imaginary components of FFT coefficients were analyzed to test consistency with Gaussian statistics expected from stationary noise.

Radiometer behavior was investigated by averaging power spectra computed from multiple independent data blocks. The reduction in variance of the averaged power spectrum was quantified as a function of the number of averaged blocks and compared to the expected 1/√N scaling. Autocorrelation functions were computed directly in the time domain and compared with power spectra obtained via the Wiener–Khinchin theorem to verify consistency between time- and frequency-domain representations of noise.


# Time-domain representation

```python
import numpy as np
import matplotlib.pyplot as plt
import xavis_package as xp


# Plot colors 
COLOR_SINE    = "#B39DDB"   # lavender 
COLOR_SAMPLES = "#7FC97F"   # light green

def plot_samples_with_underlying_fit(
    filename,
    fs,
    f0,
    capture_idx=7,
    n_show=30,
    title=""):

    #Load one capture 
    data = xp.load_npz_arr0(filename)              
    x_full = np.asarray(data[capture_idx]).ravel()  # 1D trace

    # Keep only first N samples 
    N = min(len(x_full), n_show)
    x = x_full[:N]

    # Build time axis in seconds 
    t = xp.time_axis(N, fs)


    #Fit sine at frequency f0

    w = 2 * np.pi * f0

    # Linear model: x ≈ a sin(wt) + b cos(wt) + c
    # This solves for amplitude/phase implicitly (a,b) plus a DC offset (c).
    A = np.column_stack([
        np.sin(w * t),
        np.cos(w * t),
        np.ones_like(t)])

    a, b, c = np.linalg.lstsq(A, x, rcond=None)[0]

    # nice continuous sine curve over the same window
    t_fine = np.linspace(t[0], t[-1], 2000)
    x_fit = a * np.sin(w * t_fine) + b * np.cos(w * t_fine) + c

    
    # Plot: fitted sine + samples
    plt.figure(figsize=(20, 6))

    # Underlying (best-fit) sine
    plt.plot(
        t_fine * 1e6, x_fit,
        color=COLOR_SINE,
        linewidth=4,
        label="Underlying signal (best-fit sine)")

    # Sampled points connected by lines
    plt.plot(t * 1e6, x,
        color=COLOR_SAMPLES,
        marker="o",
        linestyle="-",
        linewidth=4,
        label="Samples connected by lines")

    # Emphasize points
    plt.scatter(t * 1e6, x, color=COLOR_SAMPLES,s=200)

    # Axes labels / title
    plt.xlabel("Time (µs)", fontsize=18)
    plt.ylabel("Amplitude (ADC units)", fontsize=18)
    plt.title(title if title else f"{filename} (fs={fs/1e6:.2f} MHz)", fontsize=20)

    # Tick styling
    plt.tick_params(axis="both", which="major", labelsize=16)

    # Grid and legend
    plt.grid(True, alpha=0.3)
    plt.legend(loc="lower right", fontsize=16, frameon=True)

    # Match your original viewing window
    plt.xlim(0, 9)

    plt.tight_layout()
    plt.show()
    

f0 = 600e3

cases_600 = [
    ("below_lim_600khz.npz",  1.0e6, "600 kHz tone @ fs=1.0 MHz (below Nyquist)"),
    ("ny_lim_600khz.npz",     1.2e6, "600 kHz tone @ fs=1.2 MHz (Nyquist)"),
    ("above_lim_600khz.npz",  1.4e6, "600 kHz tone @ fs=1.4 MHz (above Nyquist)"),]

for filename, fs, title in cases_600:
    plot_samples_with_underlying_fit(
        filename=filename,
        fs=fs,
        f0=f0,
        capture_idx=7,
        n_show=25,   # keep consistent across plots for fair comparison
        title=title)
```

### Time-Domain Sampling and Aliasing Analysis

The 600 kHz tone is used here as a representative example. Similar time-domain behavior was observed for the 700 kHz and 800 kHz tones under their respective below, at, and above Nyquist sampling conditions, so the 600 kHz case is shown here for clarity.

Figure 1: (fs = 1.0 MHz, below Nyquist):
At 1.0 MHz the Nyquist frequency is 500 kHz, so a 600 kHz tone is above the highest frequency that can be represented without ambiguity. In sampled data, any frequency component above Nyquist “folds” back into the 0 to 500 kHz interval because sampling replicates the spectrum at integer multiples of the sampling rate. For this case, the 600 kHz tone is indistinguishable from a tone at 400 kHz because 600 kHz and 1.0 MHz − 600 kHz = 400 kHz produce the same sequence of sample values when sampled at 1.0 MHz. In the time-domain plot, this appears as points that oscillate more slowly than the true 600 kHz sinusoid, since the discrete samples track the aliased phase corresponding to 400 kHz rather than the true signal. This mismatch is not caused by noise or fitting error but rather it is deterministic information loss from undersampling.

Figure 2: (fs = 1.2 MHz, exactly Nyquist):
At 1.2 MHz the Nyquist frequency is 600 kHz, placing the sinusoid right at the boundary of the sampling theorem. The sampling interval is 0.833 microseconds, while the signal period is 1.667 microseconds, so the waveform is sampled at exactly two points per cycle. This creates a special “phase-locked” situation where successive samples alternate between the same two phase locations every period (for an ideal sinusoid with stable timing). That is why the points can appear aligned with a sine curve. However, this is a degenerate representation because two samples per period provide essentially no redundancy and there is no intermediate phase coverage to constrain what happens between sample times. Any small frequency offset, phase drift, or sampling jitter shifts those two sample phases, dramatically changing the observed sequence and potentially making the waveform look like a different sinusoid. In practice, sampling exactly at Nyquist is therefore fragile: it is the mathematical limit where reconstruction is theoretically possible but extremely sensitive to imperfections.

Figure 3: (fs = 1.4 MHz, above Nyquist):
At 1.4 MHz the Nyquist frequency is 700 kHz, so the 600 kHz tone is now safely below Nyquist and does not fold into a different frequency band. The sampling interval is shorter, giving about 2.33 samples per cycle, which means the sampled points advance through many different phases of the sinusoid rather than repeating the same two phases each period. That phase progression is what allows the discrete samples to encode the waveform shape and prevents aliasing. The points still do not land perfectly on a single repeated “visual sine trace” because the samples-per-cycle ratio is not an integer so the sampling phase walks relative to the signal from cycle to cycle. This non-repeating phase pattern is actually a sign of healthy sampling above Nyquist as it indicates that the discrete sequence contains enough varying phase information to match the true 600 kHz waveform instead of collapsing to an aliased lower-frequency signal.

# Frequency-domain representation

```python
import numpy as np
import matplotlib.pyplot as plt
import xavis_package as xp

COLOR_ALIASED   = "#7FC97F"   # purple
COLOR_UNALIASED = "#B39DDB"   # green
COLOR_NYQ       = "#1F78B4"   # blue

STYLE_TRUE  = (0, (2, 2))
STYLE_ALIAS = (0, (2, 2))
STYLE_NYQ   = (0, (1, 2))


def overlay_alias_unaliased_per_tone(
    configs,
    capture_idx=7,
    fmin_hz=1.0,
    base_font=22,        
    zoom_khz=None
):
    """
    Large-format, cohesive aliasing overlay plots.
    """

    # figure style
    plt.rcParams.update({
        "font.size": base_font,
        "axes.titlesize": base_font + 6,
        "axes.labelsize": base_font + 2,
        "xtick.labelsize": base_font,
        "ytick.labelsize": base_font,
        "legend.fontsize": base_font - 7,
        "lines.linewidth": 4,
    })

    nrows = len(configs)

    # --- FIG SIZE ---
    fig, axes = plt.subplots(
        nrows,
        1,
        figsize=(22, 10 * nrows),   
        sharex=False
    )

    if nrows == 1:
        axes = [axes]

    for i, cfg in enumerate(configs):
        f0 = cfg["f0"]
        label = cfg["label"]

        # --- Aliased ---
        fsA = cfg["alias"]["fs"]
        fileA = cfg["alias"]["file"]
        xA = xp.load_npz_arr0(fileA)[capture_idx]
        fA, PA = xp.fft_spectrum(xA, fsA, one_sided=True)

        f_alias = xp.alias_frequency(f0, fsA)
        peak_alias = xp.dominant_frequency(fA, PA, fmin=fmin_hz)

        # --- Unaliased ---
        fsU = cfg["unaliased"]["fs"]
        fileU = cfg["unaliased"]["file"]
        xU = xp.load_npz_arr0(fileU)[capture_idx]
        fU, PU = xp.fft_spectrum(xU, fsU, one_sided=True)

        peak_unaliased = xp.dominant_frequency(fU, PU, fmin=fmin_hz)

        ax = axes[i]

        # --- Spectra overlays ---
        ax.plot(
            fA / 1e3, PA,
            color=COLOR_ALIASED,
            label=f"Aliased spectrum (fs = {fsA/1e6:.2f} MHz)"
        )
        ax.plot(
            fU / 1e3, PU,
            color=COLOR_UNALIASED,
            alpha=0.95,
            label=f"Unaliased spectrum (fs = {fsU/1e6:.2f} MHz)"
        )

        # --- Theory lines ---
        ax.axvline(
            f0 / 1e3,
            color=COLOR_UNALIASED,
            linestyle=STYLE_TRUE,
            linewidth=4.0,
            label=f"True frequency f₀ = {f0/1e3:.0f} kHz"
        )
        ax.axvline(
            f_alias / 1e3,
            color=COLOR_ALIASED,
            linestyle=STYLE_ALIAS,
            linewidth=4.0,
            label=f"Aliased frequency = {f_alias/1e3:.0f} kHz"
        )
        ax.axvline(
            fsA / 2 / 1e3,
            color=COLOR_NYQ,
            linestyle=STYLE_NYQ,
            linewidth=4.0,
            label="Nyquist frequency (fs/2)"
        )

        ax.set_yscale("log")
        ax.set_xlabel("Frequency (kHz)")
        ax.set_ylabel("Power |FFT|² arb. units)")
        ax.grid(True, which="both", alpha=0.25)

        # --- Consistent zoom per tone ---
        if np.isclose(f0, 600e3):
            ax.set_xlim(300, 700)
        elif np.isclose(f0, 800e3):
            ax.set_xlim(500, 900)
        elif zoom_khz is not None:
            ax.set_xlim(zoom_khz)

        ax.set_title(
            f"{label} — aliased peak ≈ {peak_alias/1e3:.0f} kHz, "
            f"unaliased peak ≈ {peak_unaliased/1e3:.0f} kHz",
            pad=14
        )

        # --- Legend ---
        leg = ax.legend(
            loc="lower right",
            frameon=True,
            framealpha=0.75,
            facecolor="white",
            edgecolor="0.2"
        )
        for line in leg.get_lines():
            line.set_linewidth(5.0)

        # --- Console diagnostics ---
        print(f"{label}:")
        print(f"  THEORY alias frequency: {f_alias/1e3:.0f} kHz")
        print(f"  MEASURED aliased peak: {peak_alias/1e3:.0f} kHz")
        print(f"  MEASURED unaliased peak: {peak_unaliased/1e3:.0f} kHz")


    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()



configs = [
    {
        "f0": 600e3,
        "label": "600 kHz tone",
        "alias": {"fs": 1.0e6, "file": "below_lim_600khz.npz"},
        "unaliased": {"fs": 1.4e6, "file": "above_lim_600khz.npz"},
    },
]

overlay_alias_unaliased_per_tone(configs, capture_idx=1, base_font=30)
```

<div style="text-align: center; margin-top: 10px;">

<b>Figure 4.</b> Overlay of aliased and unaliased power spectra for a 600 kHz sinusoidal input, illustrating deterministic aliasing in the frequency domain. The green curve shows the spectrum sampled at 1.0 MHz, where the signal lies above Nyquist and appears at an aliased frequency of approximately 400 kHz, while the purple curve shows the unaliased spectrum sampled at 1.4 MHz with a peak at the true 600 kHz frequency. Green vertical line marks the true signal frequency, the purple vertical line marks the predicted unalias frequency, and the blue line marks the Nyquist frequency for the aliased case. 
    
Because the spectrum is plotted as power, proportional to the squared magnitude of the Fourier transform, the dominant peaks correspond to the frequencies where the signal’s energy is concentrated, revealing that undersampling shifts this energy to the aliased frequency while preserving its total spectral power.
    
This comparison provides a quantitative confirmation of Nyquist aliasing in the frequency domain. For a 600 kHz input sampled at 1.0 MHz, the predicted alias frequency is |f₀ − fₛ| = 400 kHz, which agrees exactly with the measured spectral peak. When the same signal is sampled above Nyquist at 1.4 MHz, the dominant peak appears at the true signal frequency of 600 kHz, confirming that the frequency shift observed in the undersampled case is not due to distortion or noise but arises from deterministic spectral folding. The similar peak power observed in both cases further demonstrates that aliasing redistributes spectral energy rather than attenuating it, reinforcing the interpretation of aliasing as a predictable consequence of sampling theory.

# Voltage and Power Spectra for 600 kHz 

```python
import numpy as np
import matplotlib.pyplot as plt
import xavis_package as xp


# Global color palette (consistent across report)
COLOR_MAG   = "#B39DDB"  # lavender (|X|)
COLOR_POWER = "#B39DDB"  # lavender (|X|^2)
COLOR_IMAG  = "#7FC97F"  # light green (Im{X})


def get_two_sided_spectrum(x, fs, remove_mean=True, window=True, pad_to=None, zero_dc=True):

    x = np.asarray(x, dtype=float).ravel()

    # Remove mean before windowing 
    if remove_mean:
        x = x - np.mean(x)

    N = len(x)

    # (reduces leakage)
    if window:
        w = np.hanning(N)
        x = x * w

        if remove_mean:
            x = x - np.mean(x)

    # zero-padding
    nfft = pad_to if (pad_to is not None and pad_to > N) else N

    # FFT and frequency axis
    X = np.fft.fft(x, n=nfft)
    f = np.fft.fftfreq(nfft, d=1/fs)

    # Optional: remove the DC bin entirely
    if zero_dc:
        X[0] = 0

    return f, X


def plot_spectra_grids_with_imag_zoom_cohesive(
    cases,
    capture_idx=7,
    pad_to=4096,
    base_font=18,
    imag_zoom_percentile=99.5,
    xlims_vp_by_fs_khz=None,
    xlims_imag_by_fs_khz=None,):
    

    if xlims_vp_by_fs_khz is None:
        xlims_vp_by_fs_khz = {}
    if xlims_imag_by_fs_khz is None:
        xlims_imag_by_fs_khz = {}

    # Consistent style scaling
    plt.rcParams.update({
        "font.size": base_font,
        "axes.titlesize": base_font + 6,
        "axes.labelsize": base_font + 1,
        "xtick.labelsize": base_font - 2,
        "ytick.labelsize": base_font - 2,
        "legend.fontsize": base_font - 2,
        "lines.linewidth": 4,})

    nrows = len(cases)

    def get_xlim(fs, xlim_dict):
        return xlim_dict.get(fs, (-fs/2/1e3, fs/2/1e3))


    # FIGURE 1: |X(f)| and |X(f)|^2
    fig1, axes1 = plt.subplots(
        nrows=nrows,
        ncols=2,
        figsize=(20, 7 * nrows),
        sharex="row",
        constrained_layout=True)
    
    if nrows == 1:
        axes1 = np.array([axes1])

    axes1[0, 0].set_title("Voltage Spectrum Magnitude  |X(f)|", pad=12)
    axes1[0, 1].set_title("Power Spectrum  |X(f)|²", pad=12)

    for i, (filename, fs, row_title) in enumerate(cases):
        x = np.asarray(xp.load_npz_arr0(filename)[capture_idx]).ravel()

        f, X = get_two_sided_spectrum(
            x, fs,
            remove_mean=True,
            window=True,
            pad_to=pad_to,
            zero_dc=True
        )
        f_khz = f / 1e3

        mag = np.abs(X)
        power = mag**2

        xlim_khz = get_xlim(fs, xlims_vp_by_fs_khz)

        axL = axes1[i, 0]
        axR = axes1[i, 1]

        axL.plot(f_khz, mag, color=COLOR_MAG)
        axL.set_xlim(xlim_khz)
        axL.set_ylabel("|X(f)| (arb. units)")
        axL.grid(True, alpha=0.3)

        axR.plot(f_khz, power, color=COLOR_POWER)
        axR.set_yscale("log")
        axR.set_xlim(xlim_khz)
        axR.set_ylabel("|X(f)|² (arb. units)")
        axR.grid(True, alpha=0.3)

        if fs == 1.2e6:
            axL.set_ylim(0, 17000)
            axR.set_ylim(1e-2, 1e6)
            
        axL.text(
            -0.15, 0.5, row_title,
            transform=axL.transAxes,
            rotation=90,
            va="center",
            ha="center",
            fontsize=base_font + 1,
            fontweight="bold")

    axes1[-1, 0].set_xlabel("Frequency (kHz)")
    axes1[-1, 1].set_xlabel("Frequency (kHz)")
    

    fig1.suptitle(
        "600 kHz Voltage Magnitude and Power Spectra",
        fontsize=base_font + 10,
        y=1.02)
    
    plt.show()

 
    # FIGURE 2: Im{X(f)} 
    fig2, axes2 = plt.subplots(
        nrows=nrows,
        ncols=1,
        figsize=(20, 7 * nrows),
        constrained_layout=True)
    if nrows == 1:
        axes2 = np.array([axes2])

    # Precompute imag arrays to set one global y-scale
    imag_all = []
    f_all = []

    for (filename, fs, _) in cases:
        x = np.asarray(xp.load_npz_arr0(filename)[capture_idx]).ravel()
        f, X = get_two_sided_spectrum(
            x, fs,
            remove_mean=True,
            window=True,
            pad_to=pad_to,
            zero_dc=True )
        imag_all.append(X.imag)
        f_all.append(f)

    imag_concat = np.concatenate([np.abs(im) for im in imag_all])
    y_max = np.percentile(imag_concat, imag_zoom_percentile)
    if y_max == 0:
        y_max = np.max(imag_concat) + 1e-12

    for i, (filename, fs, row_title) in enumerate(cases):
        f_khz = f_all[i] / 1e3
        imag = imag_all[i]

        ax = axes2[i]
        ax.plot(f_khz, imag, color=COLOR_IMAG)
        ax.set_xlim(get_xlim(fs, xlims_imag_by_fs_khz))
        ax.set_ylim(-y_max, y_max)
        ax.set_ylabel("Im{X(f)}")
        ax.set_title(row_title, fontsize=base_font + 3, pad=8)
        ax.grid(True, alpha=0.3)

    axes2[-1].set_xlabel("Frequency (kHz)")

    fig2.suptitle(
        "600 kHz Imaginary Component of the Spectrum",
        fontsize=base_font + 10,
        y=1.02)
    plt.show()


# Frequency-axis zoom windows (kHz), keyed by sampling rate
xlims_vp = {
    1.0e6: (-450, 450),
    1.2e6: (-650, 650),
    1.4e6: (-650, 650),
    3.2e6: (-900, 900),}

xlims_im = {
    1.0e6: (-450, 450),
    1.2e6: (-500, 650),
    1.4e6: (-650, 650),
    3.2e6: (-900, 900),}

plot_spectra_grids_with_imag_zoom_cohesive(
    cases_600,
    capture_idx=1,
    pad_to=4096,
    base_font=15,
    imag_zoom_percentile=99.5,
    xlims_vp_by_fs_khz=xlims_vp,
    xlims_imag_by_fs_khz=xlims_im)
```

**Figure 4.** Voltage magnitude spectra, power spectra, and imaginary components of the Fourier transform for a 600 kHz sinusoidal input sampled at three different rates relative to the Nyquist limit.  

For the upper panel, the left column shows the two-sided voltage spectrum magnitude |X(f)|, while the right column shows the corresponding power spectrum |X(f)|² plotted on a logarithmic scale. Rows correspond to sampling below Nyquist (fₛ = 1.0MHz), at Nyquist (fₛ = 1.2MHz), and above Nyquist (fₛ = 1.4MHz).  

The lower panel shows the imaginary component of the Fourier transform, Im{X(f)}, for the same three sampling conditions, illustrating the phase structure of the sampled signal in the frequency domain.

### Voltage and Power Spectra Analysis

#### Below Nyquist (fₛ = 1.0 MHz)

When the 600 kHz tone is sampled at 1.0 MHz, the Nyquist frequency is 500 kHz, placing the signal above the maximum frequency that can be uniquely represented. In the voltage spectrum magnitude, the dominant peaks appear at approximately ±400 kHz rather than at the true signal frequency. This behavior is consistent with the predicted alias frequency (f_alias = |f₀ − fₛ| = 400 kHz).

The appearance of peaks at both +400 kHz and −400 kHz occurs because the recorded time-domain signal is real-valued, and the Fourier transform of any real signal is symmetric. As a result, any frequency component present at a positive frequency must also appear at the corresponding negative frequency, even though they represent the same physical oscillation.

The corresponding power spectrum reinforces that the power is proportional to the squared magnitude of the Fourier transform. The symmetric peaks therefore reflect the mathematical properties of the Fourier transform rather than the presence of two distinct signals. This redistribution of spectral energy demonstrates that undersampling does not attenuate the signal but instead relocates its energy to a lower frequency band, confirming that the observed behavior arises from deterministic aliasing rather than noise or distortion.


---

#### At Nyquist (fₛ = 1.4 MHz)

At a sampling rate of 1.2 MHz, the Nyquist frequency is exactly 600 kHz, placing the signal at the theoretical boundary of the sampling theorem. In this case, the voltage spectrum shows a sharp feature at the Nyquist frequency, and the power spectrum displays a dominant peak precisely at 600 kHz. No spectral content appears beyond 600 kHz because, by definition, frequencies above the Nyquist limit cannot be uniquely represented in sampled data and are folded back into the allowable frequency range.

Although the frequency content appears correctly placed, this configuration represents a sampling condition. With only two samples per cycle, the spectrum contains no redundancy, and small deviations in signal frequency, phase, or timing would significantly alter the observed result. The power spectrum therefore appears well-defined, but the sampling configuration is fragile and highly sensitive to imperfections, illustrating why sampling exactly at Nyquist is generally avoided in practice.

---

#### Above Nyquist (fₛ = 1.4 MHz)

When the signal is sampled above Nyquist, the voltage spectrum magnitude shows clear peaks at ±600 kHz, corresponding to the true signal frequency and its symmetric counterpart required for a real value time-domain signal. The power spectrum also exhibits dominant peaks at these same frequencies, with no evidence of spectral folding into lower frequencies, confirming that the signal frequency is correctly represented when sampling rate exceeds twice the signal frequency.

In this regime, the sampling interval provides more than two samples per cycle, allowing the discrete-time signal to encode sufficient phase information to uniquely represent the waveform. The symmetric appearance of the spectral peaks reflects the real-valued nature of the recorded time-domain data rather than aliasing, and the absence of energy at aliased frequencies confirms that the signal is being sampled correctly.

---

### Imaginary Component Analysis

The imaginary component of the Fourier transform provides insight into the phase structure of the sampled signal.

Below Nyquist, strong imaginary components appear at the aliased frequencies, reflecting the phase evolution associated with folded spectral content. At the Nyquist rate, the imaginary component collapses to a narrow feature at the Nyquist frequency, consistent with the phase-locked, two-sample-per-cycle sampling condition observed in the time-domain analysis. 

At Nyquist, a small residual feature is also visible near 0 Hz, which arises from slight DC offsets and numerical effects introduced by finite sampling and windowing; because the signal is sampled at exactly two points per cycle, even minor asymmetries can produce a nonzero low frequency component in the imaginary spectrum.

Above Nyquist, the imaginary component exhibits pronounced features at the true signal frequency and its symmetric counterpart, indicating well-resolved phase information across the spectrum. The clean structure of Im{X(f)} in this regime further supports the conclusion that sampling above Nyquist preserves both amplitude and phase information, enabling accurate spectral interpretation.

```python
import numpy as np
import matplotlib.pyplot as plt
import xavis_package as xp


COLOR_NOISY = "#B39DDB"     # noisy (lavender)
COLOR_CLEAN = "#7FC97F"     # clean (green)

# Sampling parameters
fs = 2.2e6  # sampling rate [Hz]
ch = 7      # channel


# Load time-domain data
clean_raw = xp.load_npz_arr0("way_above_lim_800khz.npz")[ch].astype(np.float64)
noisy_raw = xp.load_npz_arr0("Nway_above_lim_800khz.npz")[ch].astype(np.float64)

# Remove DC offset
clean = clean_raw - np.mean(clean_raw)
noisy = noisy_raw - np.mean(noisy_raw)

# Apply Hann window (RMS-normalized)
N = len(clean)
window = np.hanning(N)
window /= np.sqrt(np.mean(window**2))  # RMS normalization (keeps levels comparable)

clean_w = clean * window
noisy_w = noisy * window

# Compute one-sided spectra
f_clean, P_clean = xp.fft_spectrum(clean_w, fs, one_sided=True)
f_noisy, P_noisy = xp.fft_spectrum(noisy_w, fs, one_sided=True)


# FIX 1: Convert to "PSD-like" units by dividing by bin width df
# This makes spectra comparable across different N (and more physically interpretable).
df = f_clean[1] - f_clean[0]  # Hz (assumes same length for both)
PSD_clean = P_clean / df
PSD_noisy = P_noisy / df


# FIX 2: Normalize both curves to the same off-signal noise floor
# Use a band that excludes the tone peak.
# This is useful if your SDR gain changed between runs.
DO_FLOOR_NORMALIZE = True
floor_band_khz = (250, 500)   # pick a region away from the tone peak

if DO_FLOOR_NORMALIZE:
    band = (f_clean/1e3 >= floor_band_khz[0]) & (f_clean/1e3 <= floor_band_khz[1])

    # Use median for robustness (noise PSD is spiky)
    ref_clean = np.median(PSD_clean[band])
    ref_noisy = np.median(PSD_noisy[band])

    # Scale noisy spectrum so its off-signal floor matches clean (shape comparison)
    # (This just removes unknown overall gain differences.)
    PSD_noisy = PSD_noisy * (ref_clean / ref_noisy)

# Plot

plt.figure(figsize=(20, 10))

plt.plot(
    f_clean / 1e3,
    PSD_clean,
    color=COLOR_CLEAN,
    label="Clean signal (Hann windowed)")

plt.plot(
    f_noisy / 1e3,
    PSD_noisy,
    color=COLOR_NOISY,
    alpha=0.85,
    label="Signal with injected noise (Hann windowed)")

plt.yscale("log")
plt.xlim(100, 900)

plt.xlabel("Frequency (kHz)")
plt.ylabel("PSD (arb. units/Hz)")  # <- updated label
plt.title("Injected noise comparison (windowed, PSD-normalized)")

plt.grid(True, which="both", alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()


if DO_FLOOR_NORMALIZE:
    print(f"Floor-normalized using median PSD in {floor_band_khz[0]}–{floor_band_khz[1]} kHz band.")
print(f"df = {df:.2f} Hz")
```

A Hann window (RMS-normalized) was applied prior to the FFT to reduce spectral leakage. In both cases, a sharp peak appears near 600 kHz, corresponding to the baseband frequency produced by downconversion of the 800 kHz RF signal in the SDR. When noise is added, the broadband power floor rises across the entire band, while the signal peak remains at the same frequency. This increase in background power reduces the signal-to-noise ratio by elevating the noise level underneath and around the signal peak.



### Noise Folding, Noise Injection, and Frequency Translation Analysis

Figures 5 and 6 together illustrate how broadband noise and sampling choices affect signal interpretation in a heterodyne receiver. Figure 5 isolates the effect of sampling rate on broadband noise: lowering fₛ reduces the Nyquist frequency and causes noise power from frequencies above fₛ/2 to fold incoherently into the observable band. Unlike the aliasing of a narrowband tone, this process does not create discrete spectral features, but instead increases the total in-band noise power even when the physical noise source remains unchanged.

Figure 6 demonstrates the same behavior in the presence of a sinusoidal signal at a fixed sampling rate (fₛ = 2.2 MHz). Although the RF signal frequency is 800 kHz, the dominant spectral peak appears near 600 kHz because the SDR employs heterodyne downconversion. In a heterodyne receiver, the incoming signal is multiplied by a local oscillator (LO), translating the entire spectrum by a fixed frequency offset so that the signal of interest appears at a lower intermediate (baseband) frequency that is easier to sample and process. The observed peak therefore corresponds to the baseband frequency |f_RF − f_LO|, not to destructive aliasing caused by undersampling.

To enable a meaningful comparison of broadband noise behavior, the spectra in Figure 6 are shown in PSD-like units (power per Hz) and normalized using the median PSD measured in an off-signal band from 250–500 kHz. After this normalization, the spectrum with injected noise exhibits a higher broadband floor and larger fluctuations, reducing the contrast between the signal peak and the surrounding background. This demonstrates that injected broadband noise degrades signal clarity by elevating background power rather than by shifting the signal frequency itself.

Taken together, these results show that while heterodyne mixing and controlled sampling can be used to reposition signals within a measured band, broadband noise is inevitably folded or injected into that same band. This tradeoff motivates the use of windowing, averaging, and bandwidth control to preserve signal interpretability in practical signal-processing systems.

In the context of mixer operation, these results highlight why single-sideband (SSB) receivers are preferred over double-sideband (DSB) designs in noise-limited systems. We demonstrated the consequences of DSB-style sideband stacking and showed why SSB separation is desirable, even though explicit sideband suppression could not be directly verified from the recorded real-valued data due to the limitations mentioned in the methods.

# Spectral leakage

```python
import numpy as np
import matplotlib.pyplot as plt
import xavis_package as xp

COLOR_GREEN    = "#7FC97F"
COLOR_LAVENDER = "#B39DDB"

BASE_FONT = 27
plt.rcParams.update({
    "font.size": BASE_FONT,
    "axes.titlesize": BASE_FONT + 4,
    "axes.labelsize": BASE_FONT + 1,
    "xtick.labelsize": BASE_FONT - 2,
    "ytick.labelsize": BASE_FONT - 2,
    "legend.fontsize": BASE_FONT - 5,
    "lines.linewidth": 2.8,})

def compute_power_spectrum(x, fs, window=None, pad_to=None):
    """
    One-sided power spectrum P = |FFT|^2 with optional window and zero-padding.
    """
    x = np.asarray(x, dtype=np.float64)
    x = x - np.mean(x)
    N = len(x)

    # Build a window matched to the segment length
    if window is None:
        w = np.ones(N)
    else:
        w = np.asarray(window, dtype=np.float64)
        if len(w) != N:
            raise ValueError(f"Window length {len(w)} must match data length {N}")

    xw = x * w

    # RMS-normalize so levels are comparable across windows
    w_rms = np.sqrt(np.mean(w**2))
    if w_rms > 0:
        xw = xw / w_rms

    # Zero-pad for smoother display (does not change true resolution)
    if pad_to is None:
        pad_to = N
    pad_to = max(pad_to, N)

    xw = np.pad(xw, (0, pad_to - N), mode="constant")

    X = np.fft.rfft(xw)
    P = np.abs(X)**2
    f = np.fft.rfftfreq(len(xw), d=1/fs)

    return f, P


def plot_leakage_window_comparison(
    fname="way_above_lim_800khz.npz",
    fs=3.2e6,
    f0=800e3,
    capture_idx=7,
    Nseg=4096,          # requested segment length (will be clipped to available)
    pad_to=262144,
    zoom_khz=(520, 680),
):
    """
    Spectral leakage demo: rectangular vs Hann on the SAME data segment.
    Uses peak-normalized dB to highlight sidelobes.
    """
    x_full = xp.load_npz_arr0(fname)[capture_idx].astype(np.float64).ravel()

    # Clip Nseg to the available samples (prevents broadcast errors)
    Nuse = min(Nseg, len(x_full))
    xseg = x_full[:Nuse]

    # Create a Hann window matched to the actual segment length
    hann = np.hanning(Nuse)

    fR, PR = compute_power_spectrum(xseg, fs, window=None,  pad_to=pad_to)
    fH, PH = compute_power_spectrum(xseg, fs, window=hann, pad_to=pad_to)

    PR_db = 10*np.log10(PR / np.max(PR))
    PH_db = 10*np.log10(PH / np.max(PH))

    plt.figure(figsize=(20, 8))
    plt.plot(fR/1e3, PR_db, color=COLOR_GREEN,
             label=f"Rectangular window (N={Nuse})")
    plt.plot(fH/1e3, PH_db, color=COLOR_LAVENDER,
             label=f"Hann window (N={Nuse})")

    plt.axvline(f0/1e3, color="0.4", linestyle="--", linewidth=2.0, alpha=0.8,
                label=f"Tone f0 = {f0/1e3:.0f} kHz")

    plt.xlim(*zoom_khz)
    plt.ylim(-120, 5)

    plt.xlabel("Frequency (kHz)")
    plt.ylabel("Relative power (dB)")
    plt.title("Spectral leakage comparison: Rectangular vs Hann window")
    plt.grid(True, which="both", alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()

    print(f"Used N = {Nuse} samples (requested Nseg={Nseg}, available={len(x_full)})")


plot_leakage_window_comparison(
    fname="way_above_lim_800khz.npz",
    fs=3.2e6,
    f0=800e3,
    capture_idx=7,
    Nseg=4096,       # fine: will clip to 2048 automatically if that's all you have
    pad_to=262144,
    zoom_khz=(750, 950),)
```

**Figure 7.** Comparison of spectral leakage for a single-tone measurement using a rectangular window and a Hann window, computed from the same \(N = 2048\) sample segment at fₛ = 3.2 MHz. The spectra are peak-normalized and shown in dB so that sidelobe structure can be compared directly. The green curve (rectangular window) exhibits strong sidelobes, indicating significant spectral leakage that spreads tone power into nearby frequencies. The lavender curve (Hann window) shows greatly suppressed sidelobes at the cost of a broader main lobe, producing a cleaner baseline for detecting weak signals near the tone.

### Leakage Analysis

The dominant peak appears near ≈ 850 kHz rather than exactly at f₀ = 800 kHz because the signal has been heterodyne downconverted within the SDR. As a result, the plotted frequency corresponds to the baseband offset |f_RF − f_LO| rather than the original RF frequency. Finite FFT resolution (Δf ≈ fₛ / N), combined with window-induced main-lobe broadening, can also shift the apparent peak location slightly. This behavior is expected and does not indicate distortion of the signal itself.

This figure closely follows predictions from Fourier theory. Sampling a finite-length sinusoid is equivalent to multiplying the signal by a window in the time domain, which shapes the frequency-domain response. A rectangular window produces a sinc-like spectrum with a narrow central peak and prominent sidelobes, while a Hann window smooths the time-domain edges and strongly suppresses sidelobes at the cost of a slightly wider main lobe. The observed reduction in leakage away from the peak when using the Hann window directly confirms this theoretical behavior.

In the context of the lab’s central question, this result highlights the practical role of windowing alongside heterodyne mixing and purposeful sampling choices. Even in the absence of injected noise, spectral leakage from a strong tone can spread power into neighboring frequency bins and complicate interpretation. By reducing sidelobe levels, the Hann window improves spectral clarity and helps ensure that frequency-domain structure reflects the true signal rather than artifacts of finite-time sampling.

# Noise statistics (Gaussianity)

```python
import numpy as np
import matplotlib.pyplot as plt
import xavis_package as xp


COLOR_GREEN    = "#7FC97F"   # measured data
COLOR_LAVENDER = "#B39DDB"   # model / reference
COLOR_PINK     = "#C9184A"   # residuals

BASE_FONT = 27
plt.rcParams.update({
    "font.size": BASE_FONT,
    "axes.titlesize": BASE_FONT + 4,
    "axes.labelsize": BASE_FONT + 1,
    "xtick.labelsize": BASE_FONT - 2,
    "ytick.labelsize": BASE_FONT - 2,
    "legend.fontsize": BASE_FONT - 5,
    "lines.linewidth": 2.8,})

fname = "Nway_above_lim_800khz.npz"
data = xp.load_npz_arr0(fname)             # expected shape: (10, Nsamples)

# Sum channels 0–9 into a single 1D noise vector
noise = np.sum(data[0:10].astype(np.float64), axis=0)

# Remove DC offset
noise = noise - np.mean(noise)


# Compute Vrms and Gaussian model (sigma = Vrms)
Vrms = np.sqrt(np.mean(noise**2))          # for zero-mean noise, Vrms ~ std
sigma = Vrms

print(f"Combined (sum of channels 0–9) Vrms = {Vrms:.4g} ADC units")

# PMF from quantized values
values, counts = np.unique(noise, return_counts=True)
pmf = counts / counts.sum()

# Gaussian PDF evaluated on a fine grid
x_grid = np.linspace(values.min(), values.max(), 2000)
gaussian_pdf = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-(x_grid**2)/(2*sigma**2))


# Residuals: PMF − Gaussian (binned probability)

dx = 1.0  # ADC quantization step
gaussian_pmf = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-(values**2)/(2*sigma**2)) * dx
residuals = pmf - gaussian_pmf


# Plot: PMF (bars) vs Gaussian PDF (curve) + residuals

plt.figure(figsize=(20, 10))

# PMF for quantized noise
plt.bar(
    values,
    pmf,
    width=1.0,
    color=COLOR_GREEN,
    alpha=0.55,
    label="Noise PMF")

# Gaussian PDF using sigma = Vrms
plt.plot(
    x_grid,
    gaussian_pdf,
    color=COLOR_LAVENDER,
    linewidth=7,
    label=r"Gaussian Curve $V_{\mathrm{RMS}}$")

# Residuals
plt.scatter(
    values,
    residuals,
    color=COLOR_PINK,
    s=120,
    zorder=3,
    label="Residuals (PMF − Gaussian)")

plt.axhline(0, color="0.4", linestyle="--", linewidth=1.5, alpha=0.6)

plt.xlabel("Voltage ")
plt.ylabel("Probability density")
plt.title("Noise distribution vs Gaussian model")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

**Figure 8.** Distribution of measured broadband noise compared to a Gaussian model. The green bars show the empirical probability mass function (PMF) of the noise voltage, obtained by summing channels 0–9 and removing the DC offset. The purple curve shows a zero-mean Gaussian probability density function with standard deviation σ = V₍RMS₎, computed directly from the data. Red points indicate the residuals (PMF − Gaussian model) at each quantized voltage level, highlighting small deviations between the measured distribution and the ideal Gaussian expectation.

**Analysis:** This figure demonstrates that the measured noise closely follows a Gaussian distribution, as expected for broadband noise produced by many independent random processes and further reinforced by summing multiple channels. The strong agreement between the PMF and the Gaussian model supports the assumption that the injected noise can be treated as statistically Gaussian in subsequent spectral and autocorrelation analyses. The residual structure is small and symmetric, with minor deviations attributable to finite sample size, ADC quantization, and non-ideal analog front-end effects. 

This figure supports the radiometer result discussed later in the report. The radiometer equation predicts that the uncertainty in measured power decreases as 1/√N when averaging N independent samples, which assumes the noise is Gaussian. The close match between the measured noise distribution and the Gaussian model in Figure 8 shows that this assumption is reasonable for the injected broadband noise. Since the noise fluctuations are well behaved and statistically random, averaging multiple FFT segments reduces the noise variance as expected, explaining the observed improvement in spectral stability with increasing averages.



# Radiometer behavior (averaging)

```python
fs_ref = 3.2e6
x_ref = xp.load_npz_arr0("Nway_above_lim_800khz.npz")[0].astype(np.float64)
x_ref -= np.mean(x_ref)

N_total = len(x_ref)

block_size = 512 if N_total >= 512 else max(64, N_total // 4)
max_blocks = N_total // block_size

Ns_requested = [1, 2, 4, 8, 16]
Ns = [n for n in Ns_requested if n <= max_blocks]

plt.figure(figsize=(20, 8))

for N_avg in Ns:
    linestyle = "-"   # ensure always defined

    x_use = x_ref[:N_avg * block_size]
    blocks = x_use.reshape(N_avg, block_size)

    P_list = []
    for b in blocks:
        f, P = xp.fft_spectrum(b, fs_ref, one_sided=True)
        P_list.append(P)

    P_avg = np.mean(P_list, axis=0)

    if N_avg == 1:
        color = COLOR_LAVENDER
        alpha = 0.6
    elif N_avg == 2:
        color = COLOR_GREEN
        alpha = 0.8
    else:
        color = COLOR_PINK
        alpha = 1.0

    plt.plot(f / 1e3, P_avg, color=color, alpha=alpha,
             linestyle=linestyle, label=f"N = {N_avg} averages")


plt.yscale("log")
plt.xlim(0,1200)
plt.ylim(0)
plt.xlabel("Frequency (kHz)")
plt.ylabel("Power |FFT|² (arb. units)")
plt.title("Noise PSD reduction through averaging (radiometer behavior)")
plt.grid(True, which="both", alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

**Figure 9.**  
Power spectral density (PSD) of broadband noise sampled at fₛ = 3.2 MHz, shown for different numbers of averaged FFT segments (N = 1, 2, and 4). Each curve represents the mean PSD obtained by averaging N independent FFTs of equal-length time-domain blocks. The PSD is plotted on a logarithmic scale to emphasize changes in statistical fluctuations across frequency. As N increases, the spectrum appears progressively smoother while maintaining a similar overall noise level.

**Analysis.**  
This figure demonstrates the effect of averaging on broadband noise statistics. For a single FFT (N = 1), the PSD exhibits large bin-to-bin fluctuations due to the random nature of noise. As additional independent FFTs are averaged (N = 2 and N = 4), these fluctuations decrease, producing a smoother spectrum without shifting the mean noise level. This behavior is consistent with the radiometer relation, where the uncertainty in measured noise power decreases approximately as 1/√N.

In the context of this experiment, averaging improves signal interpretability by reducing noise variance rather than removing noise power. While aliasing and heterodyne mixing determine where noise and signals appear in frequency space, averaging provides a practical way to suppress random fluctuations and make underlying spectral features easier to identify.

**Sampling choices and experimental control.**  
A relatively high sampling rate (fₛ = 3.2 MHz) was used for the noise averaging analysis to ensure that the measured noise behavior reflects statistical effects rather than artifacts of undersampling. This allows changes in the measured PSD to be attributed primarily to averaging and noise statistics rather than frequency folding.

To isolate the effect of averaging, all other parameters were held constant across the measurements: the same noise source, identical SDR configuration, fixed block size, and consistent FFT processing were used for each value of N. By varying only the number of averaged segments, the observed reduction in spectral fluctuations can be directly linked to the 1/√N scaling predicted by the radiometer equation.

Operating at a sufficiently high sampling rate and applying averaging demonstrates how noise can be statistically suppressed without altering signal frequency content, enabling cleaner signal interpretation even in noise-limited conditions. By first characterizing noise behavior at a sampling rate where aliasing is negligible, and then separately examining cases where aliasing is present, the experiment separates deterministic frequency folding from statistical noise fluctuations. This allows aliasing to be treated as a deliberate signal-processing tool rather than a source of ambiguity.

```python
import numpy as np
import matplotlib.pyplot as plt
import xavis_package as xp

COLOR_GREEN    = "#7FC97F"   # measured
COLOR_LAVENDER = "#B39DDB"   # reference
COLOR_PINK     = "#C9184A"   # emphasis / fit

BASE_FONT = 27
plt.rcParams.update({
    "font.size": BASE_FONT,
    "axes.titlesize": BASE_FONT + 4,
    "axes.labelsize": BASE_FONT + 1,
    "xtick.labelsize": BASE_FONT - 2,
    "ytick.labelsize": BASE_FONT - 2,
    "legend.fontsize": BASE_FONT - 5,
    "lines.linewidth": 4,})


fname  = "Nway_above_lim_800khz.npz"
ch     = 7
fₛ     = 3.2e6   # Hz

x = xp.load_npz_arr0(fname)[ch].astype(np.float64)
x -= np.mean(x)

N_total    = len(x)
block_size = 512 if N_total >= 512 else max(64, N_total // 4)
max_blocks = N_total // block_size

Ns_requested = [1, 2, 4, 8, 16, 32]
Ns = [n for n in Ns_requested if n <= max_blocks]

print(f"Using file: {fname}, channel: {ch}, fₛ = {fₛ/1e6:.2f} MHz")
print(f"N_total = {N_total}, block_size = {block_size}, max_blocks = {max_blocks}")
print(f"Ns used = {Ns}")


# Choose ONE frequency bin in an OFF-SIGNAL band (noise-only)
# (Pick a band that avoids the tone region in your spectra)
f_test, P_test = xp.fft_spectrum(x[:block_size], fₛ, one_sided=True)

# Example: choose ~500 kHz as a "noise-only" bin (edit if needed)
f_target_hz = 500e3
idx = np.argmin(np.abs(f_test - f_target_hz))
f_bin = f_test[idx]
print(f"Using noise-only frequency bin: {f_bin/1e3:.2f} kHz")


# Compute σ(P_avg) across independent groups

sigma_vs_N = []
Ns_used = []

for N_avg in Ns:
    n_groups = max_blocks // N_avg
    if n_groups < 2:
        continue

    Pbin_groups = []

    for g in range(n_groups):
        start = g * N_avg * block_size
        stop  = start + N_avg * block_size
        blocks = x[start:stop].reshape(N_avg, block_size)

        P_list = []
        for b in blocks:
            _, P = xp.fft_spectrum(b, fₛ, one_sided=True)
            P_list.append(P)

        P_avg = np.mean(P_list, axis=0)
        Pbin_groups.append(P_avg[idx])

    sigma_vs_N.append(np.std(Pbin_groups, ddof=1))
    Ns_used.append(N_avg)

Ns_used = np.array(Ns_used, dtype=float)
sigma_arr = np.array(sigma_vs_N)

inv_sigma_arr = 1.0 / sigma_arr

# Fit slope in log–log space (expect ≈ +0.5)
slope, intercept = np.polyfit(np.log10(Ns_used), np.log10(inv_sigma_arr), 1)


# Plot

plt.figure(figsize=(16, 10))

plt.plot(
    Ns_used,
    inv_sigma_arr,
    "o-",
    color=COLOR_GREEN,
    label=r"Measured stability $1/\sigma(P_{\mathrm{avg}})$")

plt.xscale("log")
plt.yscale("log")

plt.xlabel("Number of averaged blocks N")
plt.ylabel(r"PSD stability  $1/\sigma(P_{\mathrm{avg}})$  (arb. units)")
plt.title(
    f"Radiometer behavior at {f_bin/1e3:.1f} kHz (noise-only):  "
    f"$1/\\sigma \\propto N^{{{slope:.2f}}}$  (expected +0.5)")

plt.grid(True, which="both", alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

print(f"Measured scaling: 1/sigma(P_avg) ∝ N^{slope:.2f}   (expected +0.50)")
```

**Figure 10. Radiometer behavior measured from noise PSD averaging.**  
This figure shows the measured stability of the noise power spectral density (PSD) as a function of the number of averaged FFT blocks, N, using noise data recorded at a sampling rate of fₛ = 3.20 MHz. The analysis was performed on the dataset `Nway_above_lim_800khz.npz` (channel 7), with a total of N_total = 2048 samples divided into blocks of size 512, allowing up to four independent averages. PSD stability is quantified as 1/σ(P_avg), where σ is the standard deviation of the averaged PSD measured at a representative frequency bin within the passband. The data points correspond to N = 1, 2, and 4 averaged blocks, and a power-law fit is shown on log–log axes.

**Analysis.**  
This figure probes the statistical behavior predicted by the radiometer equation, which states that the uncertainty in measured noise power decreases as 1/√N when averaging independent measurements. In this dataset, the measured scaling, 1/σ(P_avg) ∝ N^1.38, is steeper than the ideal N^0.5 expectation. This discrepancy is not unexpected given that the assumptions of the radiometer equation are not fully satisfied: only a small number of independent averages are available, FFT blocks are short, and correlations are introduced by finite windowing, spectral leakage, and the non-ideal bandpass response of the SDR front end.

The radiometer behavior was evaluated at a frequency near 500 kHz rather than at the signal peak to avoid contamination from deterministic signal structure and spectral leakage. Measuring stability in an off-signal region ensures that the observed variance reduction is driven by random noise fluctuations alone, which is the regime in which the radiometer equation applies. This choice allows the effects of averaging to be isolated from frequency translation or aliasing effects associated with the signal itself.

Despite the deviation from the ideal scaling, the key qualitative behavior is clearly observed, as increasing the number of averages systematically improves PSD stability without altering the frequency placement of spectral features. This demonstrates that averaging suppresses statistical noise fluctuations rather than shifting signal content. In the broader context of the experiment, this result complements earlier aliasing and heterodyne analyses by showing that while sampling and mixing determine where signals and noise appear in frequency space, averaging controls how reliable those features can be interpreted. Together, these techniques support the use of purposeful aliasing to shape noise folding while preserving a clean, interpretable signal in an SDR.


## Conclusion

This experiment examined how sampling rate, aliasing, noise folding, and statistical averaging shape signal interpretation in a software-defined radio (SDR). While aliasing is often treated as a failure mode of sampling, our results show that it can instead be used deliberately as a predictable frequency-translation tool, provided that its effects on broadband noise are carefully managed. This directly addresses the working question posed in the introduction: **how purposeful aliasing can be used to shape noise folding while preserving an interpretable signal**.

Time-domain and frequency-domain measurements of sinusoidal tones at 600 kHz, 700 kHz, and 800 kHz confirmed the deterministic nature of aliasing predicted by sampling theory. When signals were sampled below the Nyquist limit, their spectral energy folded to precisely the frequencies predicted by **f_alias = |f₀ − kfₛ|**, while sampling above Nyquist preserved the true signal frequency. These results agree closely with theoretical expectations and demonstrate that aliasing redistributes spectral energy rather than destroying it. Sampling exactly at the Nyquist rate produced fragile, degenerate representations, reinforcing why real systems avoid operating at this limit.

Noise measurements revealed a complementary effect, where lowering the sampling rate causes broadband noise from frequencies above **fₛ / 2** to fold incoherently into the observable band, raising the in-band noise floor without introducing discrete features. This behavior matches theoretical expectations for white or broadband noise and explains why aliasing trades frequency translation for increased noise. By comparing noise power spectral densities (PSDs) at different sampling rates while holding all other parameters fixed, the experiment isolated noise folding as a direct consequence of sampling choice rather than changes in the noise source itself.

Statistical analysis of the noise showed that its voltage distribution is well approximated by a Gaussian, validating assumptions underlying the radiometer equation. Averaging multiple FFT segments reduced variance in the measured power spectrum, consistent with the expected **1 / √N** behavior. Deviations from the ideal scaling were observed due to limited sample size, finite FFT length, and correlations introduced by windowing and the SDR front-end response. These deviations highlight practical limitations that arise in real instruments but do not undermine the qualitative conclusion that averaging improves spectral stability without altering frequency content.

Several limitations affect the interpretation of these results. The recorded data consisted of real-valued voltage samples rather than full complex I/Q streams, preventing direct verification of sideband suppression and limiting explicit demonstration of single-sideband (SSB) operation. The number of independent averages available for radiometer analysis was also small, leading to noisy scaling estimates. In addition, the exact analog bandpass response of the SDR and mixer was not independently characterized, which likely contributed to residual spectral structure.

Future work could improve on these limitations by recording and analyzing full complex I/Q data to directly demonstrate SSB versus DSB behavior, increasing data volume to better test radiometer scaling, and explicitly characterizing the analog front-end response. Applying controlled digital filtering alongside purposeful aliasing would further clarify how noise folding can be shaped in practice. Despite these limitations, the present results demonstrate that aliasing, when understood and controlled, is not merely a sampling artifact but a usable signal-processing tool in noise-limited radio systems.
