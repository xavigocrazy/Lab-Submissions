# 5.4 Voltage Power Spectra

```python
#volt of 200kHz, 3e6 sample

volt_200_3e6 = np.fft.fft(data_200_3e6)

plt.figure()
plt.plot(volt_200_3e6)
plt.ylabel("Voltage")
plt.xlabel("Sampling Index")
plt.show()
```

```python
#Voltage of 200kHz, 3e6 sample
#power of 200kHz, 3e6 sample

power_200_3e6 = np.abs(np.fft.fft(data_200_3e6))**2

plt.figure()
plt.plot(power_200_3e6)
plt.ylabel("Power")
plt.xlabel("Sampling Index")
plt.show()
```

```python
#Both on logs scales now

#volt of 200kHz, 3e6 sample

volt_200_3e6 = np.fft.fft(data_200_3e6)

plt.figure()
plt.plot(volt_200_3e6)
plt.ylabel("Voltage")
plt.xlabel("Sampling Index")
plt.yscale("log")
plt.xlim(0,1000)
plt.show()
```

```python
#power of 200kHz, 3e6 sample

power_200_3e6 = np.abs(np.fft.fft(data_200_3e6))**2

plt.figure()
plt.plot(power_200_3e6)
plt.ylabel("Power")
plt.xlabel("Sampling Index")
plt.yscale("log")
plt.show()
```

```python
#plotting real and imaginary aspects of voltage spectrum

#volt of 200kHz, 3e6 sample

volt_200_3e6 = np.fft.fft(data_200_3e6)
real_200_3e6 = np.real(volt_200_3e6)

plt.figure()
plt.plot(real_200_3e6)
plt.ylabel("Voltage")
plt.xlabel("Sampling Index")
plt.yscale("log")
plt.xlim(0,1000)
plt.title("Voltage Spectrum (Real)")
plt.show()
```

```python
#plotting real and imaginary aspects of voltage spectrum

#volt of 200kHz, 3e6 sample

volt_200_3e6 = np.fft.fft(data_200_3e6)
imag_200_3e6 = np.imag(volt_200_3e6)

plt.figure()
plt.plot(imag_200_3e6)
plt.ylabel("Voltage")
plt.xlabel("Sampling Index")
plt.yscale("log")
plt.xlim(0,1000)
plt.title("Voltage Spectrum (Imaginary)")
plt.show()
```

```python
#using numpy.fft.ifft

#power of 200kHz, 3e6 sample

power_200_3e6 = np.abs(np.fft.fft(data_200_3e6))**2
power_ifft_200_3e6 = np.fft.ifft(power_200_3e6)

plt.figure()
plt.plot(power_ifft_200_3e6)
plt.ylabel("Power")
plt.xlabel("Sampling Index")
plt.yscale("log")
plt.xlim(0,500)
plt.show()
```

```python
#Computing AFC 

power_ifft_200_3e6 = np.fft.ifft(power_200_3e6)
power_afc = np.real(power_ifft_200_3e6)
power_afc /= power_afc[0]

volt_ifft_200_3e6=np.fft.ifft(volt_200_3e6)
volt_afc = np.real(volt_ifft_200_3e6)
volt_afc /= volt_afc[0]

plt.figure()
plt.plot(power_afc, color = "green")
plt.plot(volt_afc, color = "red")
plt.ylabel("Voltage")
plt.xlabel("Sampling Index")
plt.xlim(0,250)
plt.show()
```
